```python
def user_choice():
    
    choice ='WRONG'
    within_range = False
    
    while choice.isdigit() == False or within_range == False:
        
    
    
        choice = input("Please enter a number between 1 and 9: ")
        
        if choice.isdigit() == False:
            print("Sorry that is not a digit.")
            
        if choice.isdigit() == True:
            if int(choice) in range(0,10):
                within_range = True
            else:
                #print("Sorry that is number is not between 0 and 9.")
                within_range = False
        
    
    return int(choice)
```

```python
def print_board(game_status):
    #condition to check that gamestatus has length 9
    #if len(game_status != 9):
    #    break
    #else:
        board=[' ']*len(game_status)
        for i in range(0,len(game_status)):
            if game_status[i]==0:
                board[i]=' '
            if game_status[i]==1:
                board[i]='x'
            if game_status[i]==2:
                board[i]='o'
    
        print(board[0:3])
        print(board[3:6])
        print(board[6:9])
```
